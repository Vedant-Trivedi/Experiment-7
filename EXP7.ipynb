```python
# Factorial of a number
n = int(input("Enter a Number: "))

if n < 0:
  print("Factorial is not defined for negative numbers.")
elif n == 0:
  print("Factorial of 0 is 1.")
else:
  fact = 1
  i = 1
  while i <= n:
    fact *= i
    i += 1
  print("Factorial =", fact)
```

```python
#Fibonacci series using input and while loop
n = int(input("Enter the number of terms: "))
a, b = 0, 1
count = 0

if n <= 0:
    print("Please enter a positive integer")
elif n == 1:
    print("Fibonacci sequence upto", n, ":")
    print(a)
else:
    print("Fibonacci sequence:")
    while count < n:
        print(a, end=" ")
        nth = a + b
        a = b
        b = nth
        count += 1
```

```python
n = int(input("Enter a number: "))

if n < 0:
    print("Please enter a non-negative number.")
elif n == 0:
    print(0)
else:
    print("Number in reverse: ")
    while n > 0:
        digit = n % 10
        print(digit, end="")
        n //= 10
    print()
```

```python
# Palindrome check using while loop
num = int(input("Enter a number: "))
original = num
reversed = 0

while num > 0:
    digit = num % 10
    reversed = reversed * 10 + digit
    num //= 10

if original == reversed:
    print(original, "is a palindrome.")
else:
    print(original, "is not a palindrome.")
```

```python
#String & Number Palindrome
text = input("Enter a string/number: ")
text_1 = text.lower()
reversed = text_1[::-1]
if text_1 == reversed:
    print(f"'{text}' is a palindrome.")
else:
    print(f"'{text}' is not a palindrome.")
```

```python
# String & Number Palindrome [Alternate method]
text = input("Enter a string/number: ")
text_lower = text.lower()

reversed_text = ""
i = len(text_lower) - 1
while i >= 0:
    reversed_text += text_lower[i]
    i -= 1

if text_lower == reversed_text:
    print(f"'{text}' is a palindrome.")
else:
    print(f"'{text}' is not a palindrome.")
```

```python
# Program to find the index of an element in a list using a while loop
my_list = [10, 20, 30, 40, 50, 60, 70]
search_element = int(input("Enter the element to search: "))

index = 0
found = False

while index < len(my_list):
    if my_list[index] == search_element:
        print(f"Element {search_element} found at index {index}.")
        found = True
        break
    index += 1

if not found:
    print(f"Element {search_element} not found in the list.")
```

```python
n = int(input("Enter a number (n): "))

i = 1
print(f"Odd numbers from 1 to {n}:")
while i <= n:
    if i % 2 == 0:
        i += 1
        continue
    print(i)
    i += 1
```
